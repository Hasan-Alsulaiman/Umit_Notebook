```python
import pandas as pd
import numpy as np
import random
import networkx as nx
from tqdm import tqdm
import re
import matplotlib.pyplot as plt
import json



from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
# load  details
with open("data (1).json") as f:
    data = json.load(f) 
```

```python
print(data['edges'][1]['source'])
```

```python
# captture nodes in 2 separate lists
node_list_1 = []
node_list_2 = []

# tqdm is just a progress bar, so this is a normal for loop with a progress bar.
for i in tqdm(data['edges']):
#     the links in fb_links are sperated by a comma, to split them into two lists we do:
   node_list_1.append((i['source']))
   node_list_2.append((i['target']))
```

```python
print(node_list_2[:10])
```

```python
data_df = pd.DataFrame({'node_1': node_list_1, 'node_2': node_list_2})
```

```python
 data_df.head()
```

```python
# create graph
G = nx.from_pandas_edgelist(data_df, "node_1", "node_2", create_using=nx.Graph())
# plot graph
plt.figure(figsize=(10,10))

# pos = nx.random_layout(G, seed=23)
pos = nx.spring_layout(G)
d = dict(G.degree)
node_color = [20000.0 * G.degree(v) for v in G]
nx.draw(G, node_color=node_color, with_labels=False,  pos = pos, alpha = 0.6, width = 0.7, node_size=[v * 20 for v in d.values()])

plt.show()
```
